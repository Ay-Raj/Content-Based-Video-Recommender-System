```python
#using Tfidf

tfv = TfidfVectorizer(min_df=3,  max_features=None, 
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')

# Filling NaNs with empty string
df['overview'] = df['overview'].fillna('')
```

```python
tfv_matrix = tfv.fit_transform(df['overview'])
```

```python
tfv_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
sig = cosine_similarity(tfv_matrix, tfv_matrix)
```

```python
# Reverse mapping of indices and titles
indices = pd.Series(df.index, index=df['title'])
```

```python
def give_rec(title, sig=sig):
    # Get the index corresponding to title
    idx = indices[title]

    # Get the pairwsie similarity scores 
    sig_scores = list(enumerate(sig[idx]))

    # Sort the tile 
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 5 most similar videos
    sig_scores = sig_scores[1:11]

    # title indices
    video_indices = [i[0] for i in sig_scores]

    # Top 5 most similar videos
    return df['title'].iloc[video_indices]
```

```python
# Testing the recommendation system
give_rec('Eminem - Walk On Water (Audio) ft. BeyoncÃ©')
```

```python
def give_rec_category(title):
    # Get the index corresponding to title
    idx = indices[title]

    # Get the pairwsie similarity scores 
    sig_scores = list(enumerate(sig[idx]))

    # Sort the tile 
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 5 most similar videos
    sig_scores = sig_scores[1:11]

    # title indices
    video_indices = [i[0] for i in sig_scores]

    # Top 5 most similar videos
    return df['category_id'].iloc[video_indices]

def relevance_score(iterations):
  accuracy = []
  for i in range(iterations):
    id = df.loc[i].category_id
    result = give_rec_category(df['title'].loc[i]).apply(lambda x: x == id).mean()
  accuracy.append(result)
  return np.mean(accuracy)
```

```python
relevance_score(100)
```
